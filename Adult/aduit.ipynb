```python
# حساب المتوسط لعدد ساعات العمل الإسبوعية بالنسبة لكل عمل ومن ثم ترتيب النواتج تصاعدي
# groupby: 
# أن العامود المختار لديه عناصر متكررة متواجدة في سطور مختلفة مرتبطة مع كامل السطر        
# كل عنصر متكرر يكون مفتاح لبقية البيانات المرتبطة به المتواجدة في السطر ذاته
# تضع هذه الدالة البيانات ضمن مجموعات متصلة بكل مفتاح لننفذ عليها عمليات محدد كالمتوسط 
# طبعا كل عامود يتمتع بصفات معينة لذلك نستطيع تحديد العامود المطلوب لتنفيذ العملية عليه
data.groupby('workclass')['hours-per-week'].mean().sort_values(ascending=False)
```
