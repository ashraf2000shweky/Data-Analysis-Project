<h2> pivot table </h2>
<br>
<b>
من خلال تسميته يساعد على أن نجعل البيانات مرتكزة(تتمحور) على فهرس،عامود،قيم داخل الجدول 
من خلال تعين دالة تقوم بحساب عمليات على القيم التى داخل الجدول
</b>


```python
df_1 = data.pivot_table(index= 'Origin',values ='Weight',aggfunc = np.median)
df_1 
```

```python
sns.barplot(x=df_1.index,y=df_1['Weight'])
```

```python
df_2 = data.pivot_table(index= 'Type',values ='Weight',aggfunc = np.median)
df_2
```

```python
sns.barplot(x=df_2.index,y=df_2['Weight'])
```

```python
df_3 = data.pivot_table(index= 'Cylinders',values ='MPG_City',aggfunc = np.median)
df_3
```

```python
sns.barplot(x=df_3.index,y=df_3['MPG_City'])
```

```python
df_4 = data.pivot_table(index= ['Origin','Make',],values ='Horsepower',aggfunc = np.median).reset_index()
df_4
```

```python
df_5 = df_4[df_4['Origin'] == 'Asia'] # نحدد أن السيارة من أصل أسيوي
df_5['Horsepower'].mean()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=df_5['Make'],y=df_5['Horsepower'])
```

<h2> groupby  </h2>
<br>
 <b>
 أن العامود المختار لديه عناصر متكررة متواجدة في سطور مختلفة مرتبطة مع كامل السطر        
 وكل عنصر متكرر يكون مفتاح لبقية البيانات المرتبطة به المتواجدة في السطر ذاته
 تضع هذه الدالة البيانات ضمن مجموعات متصلة بكل مفتاح لننفذ عليها عمليات محدد كالمتوسط 
وطبعا كل عامود يتمتع بصفات معينة لذلك نستطيع تحديد العامود المطلوب لتنفيذ العملية عليه
   

```python
# حساب الوسط 
data.groupby('Make')['Invoice','Horsepower'].median()
```

```python
data.groupby(['Origin','Make'])['Invoice'].median().unstack() 
# unstack تجعل عناصر الفهرس الاخير أسماء لأعمدة 
```

```python
ser = data.groupby(['Origin','Make'])['Invoice'].median()
ser
```

```python
ser.index
```

```python
ser.unstack()
```
