```python
movie = ['Comedy','Action','Romance','Drama']
votes = [5,6,3,1]
```

```python
plt.pie(votes) # شكل الفطيرة
plt.show()
```

```python
plt.title('Favorite Type of Movie')
plt.pie(votes,labels=movie,labeldistance=1.1) # يمكن التعديل على مسافة الدلالات عن الشكل
plt.legend(bbox_to_anchor=(1,1)) # تغير موضع دلالات الشكل
plt.show()
```

```python
plt.title('Favorite Type of Movie')
plt.pie(votes,labels=movie,explode =[0,0.2,0,0.1] ) # إخراج قطع الفطيرة بنسبة معينة
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

```python
plt.title('Favorite Type of Movie')
plt.pie(votes,labels=movie,explode =[0,0.2,0,0.1],startangle=45 ) # تغير زاوية البدء
plt.legend(bbox_to_anchor=(1.1,1))
plt.show()
```

```python
plt.title('Favorite Type of Movie')
plt.pie(votes,labels=movie,explode =[0,0.2,0,0.1],startangle=40,
       autopct='%0.1f%%', pctdistance=0.6) # إضافة الحصة الرقمية لكل قطعة
plt.legend(bbox_to_anchor=(1.1,1))
plt.show()
```

```python
plt.title('Favorite Type of Movie',y=1.2,fontsize=15)
plt.pie(votes,labels=movie,explode =[0,0.2,0,0.1],startangle=40,
 autopct='%0.1f%%',radius=1.5,textprops= {'fontsize':12}) # تغير نصف قطر دائرة الفطيرة وحجم خطوط عناصرها
plt.legend(bbox_to_anchor=(1.1,1.1))
plt.show()
```

```python
plt.title('Favorite Type of Movie',y=1.2,fontsize=15)
plt.pie(votes,labels=movie,explode =[0,0.2,0,0.1],startangle=40,
       autopct='%0.1f%%',radius=1.5,textprops= {'fontsize':12},
       wedgeprops={'edgecolor':'k','linewidth': 0.7}) # إضافة خطوط فصل بين الحصص
plt.legend(bbox_to_anchor=(1.1,1.1))
plt.show()
```

```python
plt.title('Favorite Type of Movie',y=1.2 )
plt.pie(votes,labels=movie,explode =[0,0.2,0,0.1],startangle=40,
       autopct='%0.1f%%',radius=1.5,textprops= {'fontsize':12},
       shadow=True) # إظهار ظل
plt.legend(bbox_to_anchor=(1.1,1.1))
plt.show()
```

```python
plt.figure(figsize = (7,7),
         edgecolor = 'k', linewidth = 4 , facecolor ='lightyellow') # إضافة حواف
plt.title('Favorite Type of Movie',y=1.2,fontsize=15)
plt.pie(votes,labels=movie,
        explode =[0,0.2,0,0.1], 
        startangle=40,
        autopct='%0.1f%%', 
        textprops= {'fontsize':12},
        shadow=True)
plt.legend(bbox_to_anchor=(1.1,1.1),fontsize=13)
plt.show()
```

```python
fig,ax =  plt.subplots(nrows=1,ncols=2) #  تعريف عدة أشكال ضمن شكل واحد مفيد جداً
fig.suptitle('Iris dataset',fontsize=20)
ax[0].hist(data['SepalLengthCm']) # ذات المفاهيم السابقة بدوال مختلفة قليلأ
ax[0].set_title('SepalLengthCm',fontsize=15)
ax[0].set_xlabel('LengthCm')
ax[0].set_ylabel('count')
plt.show()
```

```python
fig,ax =  plt.subplots(nrows=1,ncols=2)
fig.suptitle('Iris dataset',fontsize=20)

ax[0].hist(data['SepalLengthCm'])
ax[0].set_title('SepalLengthCm',fontsize=15)
ax[0].set_xlabel('LengthCm')
ax[0].set_ylabel('count')


ax[1].hist(data['PetalLengthCm'],color='b')
ax[1].set_title('PetalLengthCm',fontsize=15)
ax[1].set_xlabel('LengthCm')
ax[1].set_ylabel('count')

plt.tight_layout() # إزالة تداخل العناوين بالشكل
plt.show()
```
