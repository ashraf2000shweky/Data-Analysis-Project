```python
movies = pd.read_csv('IMDB top 1000.csv')
```

```python
# أخذ 75% من البيانات بشكل عشوائي
movies_1 = movies.sample(frac=0.75,random_state=123)
movies_1
```

```python
# بقية عناصر البيانات 
movies_2 = movies.drop(movies_1.index)
```

```python
len(movies_1) + len(movies_2) == len(movies)
```

```python
x = movies_1.index.sort_values()
```

```python
# تشكل فعرس جديد
movies_1.reset_index(drop = True)
```

```python
# الفهرسة كما في البيانات الأصلية 
movies_1.index = x
```

```python
counts = movies['Genre'].value_counts()
counts
```

```python
# أكبر ثلاثة عناصر
counts.nlargest(3)
```

```python
# أصغر ثلاثة أعمدة 
counts.nsmallest(3)
```

```python
# البيانات التى تشير الى أكثر ثلاثة أنوع أفلام تكراراً
movies[movies['Genre'].isin(counts.nlargest(3).index)]
```
