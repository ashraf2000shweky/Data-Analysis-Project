```python
import pandas as pd
```

```python
data = {
    'data': ['2018-01-01 35 20736.87 Paul',
             '2017-06-015 67 10567.70 Marry',
             '2015-07-23 20 3837.90 Noel',
             '2011-05-12 56 82320.65 Joho',
             '2010-03-26 21 28737.01 Sally',
             '2016-03-19 82 80327.63 Marry',
             '2018-08-19 90 30467.92 steve'
            ]
}
```

```python
df = pd.DataFrame(data,columns=['data'])
df
```

```python
# إستخلاص الأسم 
df['Name'] = df['data'].str.extract('([A-Z]\w{0,})') 
```

```python
df
```

```python
# إستخراج السعر 
df['Salary'] = df['data'].str.extract('(\d{0,}\.\d\d)')
```

```python
df
```

```python
# إستخراج التاريخ 
df['date'] = df['data'].str.extract('(....-..-..)')
df
```
