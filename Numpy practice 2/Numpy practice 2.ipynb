```python
# معلومات إحصائية عن البيانات 
heights = arr[:,0]
print('max heights = ',heights.max(),'cm')
print('min heights = ',heights.min(),'cm')
print('mean heights = ',heights.mean(),'cm')
print('standard deviation heights = ',heights.std(),'cm')
```

```python
# معلومات إحصائية عن البيانات 
weights = arr[:,1]
print('max weights = ',weights.max(),'kg')
print('min weights = ',weights.min(),'kg')
print('mean weights = ',weights.mean(),'kg')
print('standard deviation weights = ',weights.std(),'kg')
```

```python
# إيجاد العناصر الدالة على توزيع البيانات
# حيث أن 25% من البيانات أصغر من 156
np.percentile(heights,[25,50,75])
```
