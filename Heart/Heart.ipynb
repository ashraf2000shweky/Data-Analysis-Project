```python
# هنا نريد معرفة الأعمدة التابعة لقيم مختلفة لمثلها على شكل مخطط مناسب  
cate_val = []
cont_val = []
```

```python
for column in data.columns:
    if data[column].nunique() <= 10:
        cate_val.append(column)
    else:
        cont_val.append(column)
```

```python
cate_val
```

```python
cont_val
```

```python
data.hist(cont_val,figsize=(15,10))
plt.tight_layout()
plt.show()
```
